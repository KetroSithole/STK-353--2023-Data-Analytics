## Basic Matplotlib Commands

There are two methods:
1. Functional Method (Matlab-style)
2. Object Oriented Method (usually better approach, especially with more complex graphs)

### 1. Functional Method

```python
a = np.linspace(0, 5, 11)
b = a ** 3
```

```python
a
```

```python
plt.plot(a,b,'r') # 'r' is the color red, also
```

```python
plt.plot(a,b,'r', markersize=4, linestyle=':', marker='o') # 'r' is the color red, also
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Positive Exponential Graph')
```

**Plotting multiple sets of data on same graph**

The most straight forward way is just to call `plot` multiple times:

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

**Plotting multiple sets of data on same canvas**

```python
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)    #[subplot creates multiple plot canvases]
plt.plot(x, np.sin(x),'r--')

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'g')
```

### 2. Object Oriented Method

This approach is better when dealing with more complex plots, eg. a canvas that has multiple plots on it. 

**Objects** are variables that contain data and functions (methods) that can be used to manipulate the data. eg. lists are python objects. In `my_list.extend('5')`, the `extend()` is the method.

In the object-oriented method, the plotting functions are *methods* of explicit ``Figure`` and ``Axes`` objects.
To re-create the previous plot using this style of plotting, you might do the following:

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x), 'r--')
ax[1].plot(x, np.cos(x), 'g')
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x), 'r--')
ax[1].plot(x, np.cos(x), 'g')

ax[0].set_xlabel('X_label_axes1')                  # for a Object oriented approach, we set labels like this .set_xlabel()
ax[0].set_ylabel('Y_label_axes1')
ax[0].set_title('Title 1')

ax[1].set_xlabel('X_label_axes2')
ax[1].set_ylabel('Y_label_axes2')
ax[1].set_title('Title 2')

fig.tight_layout(pad=0.5)                        # for space around a plot


```

### Saving Figures to File

One nice feature of Matplotlib is the ability to save figures in a wide variety of formats.
Saving a figure can be done using the ``savefig()`` command.
For example, to save the previous figure as a PNG file, you can run this:

```python
fig.savefig('my_figure.png')
```

To confirm that it contains what we think it contains, let's use the IPython ``Image`` object to display the contents of this file:

```python
from IPython.display import Image
Image('my_figure.png')
```

In ``savefig()``, the file format is inferred from the extension of the given filename.
Depending on what backends you have installed, many different file formats are available.
The list of supported file types can be found for your system by using the following method of the figure canvas object:

```python
fig.canvas.get_supported_filetypes()
```

Recreate the following graph:



where $y_1 = x^3$,  $y_2 = 2x(x+10)(x-10)$ and $z=\cos(2x)$ for `x = np.linspace(-10,10, 21)`

```python
x = np.linspace(-10,10, 21)
y1 = x**3
y2 = 2*x*(x+10)*(x-10)
z = np.cos(2*x)

# First create a grid of plots  [fig]
# ax will be an array of two Axes objects
fig, ax = plt.subplots(1,2,figsize=(12, 6))              # subplot creates multiple canvases...subplot(row, column, panel)

# Call plot() method on the appropriate object
ax[0].plot(x, y1, 'm')                                   # 0 is an index for the first plot
ax[0].plot(x, y2, 'b*')
ax[1].plot(x, z, 'k--')

ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Graph of Y1 and Y2')

ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
ax[1].set_title('Graph of Z')
```
