### Monte-Carlo Integration

```python
# limits of integration
a = 0
b = np.pi # gets the value of pi
N = 1000

# array of zeros of length N
ar = np.zeros(N)

# iterating over each Value of ar and filling
# it with a random value between the limits a
# and b
for i in range (len(ar)):
    ar[i] = np.random.uniform(a,b)

# variable to store sum of the functions of
# different values of x
integral = 0.0

# function to calculate the sin of a particular
# value of x
def f(x):
    return np.sin(x)

# iterates and sums up values of different functions
# of x
for i in ar:
    integral += f(i)

# we get the answer by the formula derived above
ans = (b-a)/float(N)*integral
# ans = integral*(b-a)*1/1000 (alternatively, same thing though)

# prints the solution
print ("The value calculated by monte carlo integration is {}.".format(ans))
```

### Bootstrapping methods

#### Means examples

```python
# Original dataset
data = np.array([14, 18, 20, 22, 23, 25, 27, 29, 31, 35])

# Number of bootstrap iterations
num_iterations = 10000

# Function to compute the mean of a dataset
def compute_mean(data):
    return np.mean(data)

# Bootstrap resampling and mean estimation
bootstrap_means = [compute_mean(np.random.choice(data, size=len(data), replace=True)) for _ in range(num_iterations)]

# Calculate the 95% confidence interval for the mean
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])

# Plot a histogram of the bootstrap sample means
plt.hist(bootstrap_means, bins=30, density=True, alpha=0.6)
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Bootstrap Sampling Distribution of the Mean')
plt.axvline(x=confidence_interval[0], color='r', linestyle='--', label='95% CI Lower Bound')
plt.axvline(x=confidence_interval[1], color='g', linestyle='--', label='95% CI Upper Bound')
plt.legend()
plt.show()

# Print the results
print("Bootstrap 95% Confidence Interval for the Mean:", confidence_interval)
```

#### Medians Example

```python
# Original dataset
data = np.array([14, 18, 20, 22, 23, 25, 27, 29, 31, 35])

# Number of bootstrap iterations
num_iterations = 10000

# Function to compute the median of a dataset
def compute_median(data):
    return np.median(data)

# Bootstrap resampling and median estimation
bootstrap_medians = [compute_median(np.random.choice(data, size=len(data), replace=True)) for _ in range(num_iterations)]

# Calculate the 95% confidence interval for the median
confidence_interval = np.percentile(bootstrap_medians, [2.5, 97.5])

# Plot a histogram of the bootstrap sample medians
plt.hist(bootstrap_medians, bins=30, density=True, alpha=0.6)
plt.xlabel('Sample Median')
plt.ylabel('Probability Density')
plt.title('Bootstrap Sampling Distribution of the Median')
plt.axvline(x=confidence_interval[0], color='r', linestyle='--', label='95% CI Lower Bound')
plt.axvline(x=confidence_interval[1], color='g', linestyle='--', label='95% CI Upper Bound')
plt.legend()
plt.show()

# Print the results
print("Bootstrap 95% Confidence Interval for the Median:", confidence_interval)
```

#### R-squared example

```python
# Simulate a linear dataset
np.random.seed(0)
X = np.random.rand(100) * 10
y = 2 * X + np.random.randn(100)

# Linear regression model (y = beta * X + epsilon)
def linear_regression(X, beta):
    return beta * X

# Function to compute R-squared
def compute_r_squared(y_true, y_pred):
    total_variance = np.sum((y_true - np.mean(y_true))**2)
    explained_variance = np.sum((y_pred - np.mean(y_true))**2)
    r_squared = explained_variance / total_variance
    return r_squared

# Number of bootstrap iterations
num_iterations = 1000

# Bootstrap resampling and R-squared estimation
bootstrap_r_squared = []
for _ in range(num_iterations):
    # Resample with replacement
    indices = np.random.choice(len(X), size=len(X), replace=True)
    X_resampled = X[indices]
    y_resampled = y[indices]
    
    # Fit a linear regression model to the resampled data
    beta_hat = np.sum(X_resampled * y_resampled) / np.sum(X_resampled**2)
    
    # Make predictions on the original data
    y_pred = linear_regression(X, beta_hat)
    
    # Compute R-squared and store it
    r_squared = compute_r_squared(y, y_pred)
    bootstrap_r_squared.append(r_squared)

# Calculate the 95% confidence interval for R-squared
confidence_interval = np.percentile(bootstrap_r_squared, [2.5, 97.5])

# Plot a histogram of the bootstrap R-squared values
plt.hist(bootstrap_r_squared, bins=30, density=True, alpha=0.6)
plt.xlabel('R-squared')
plt.ylabel('Probability Density')
plt.title('Bootstrap Sampling Distribution of R-squared')
plt.axvline(x=confidence_interval[0], color='r', linestyle='--', label='95% CI Lower Bound')
plt.axvline(x=confidence_interval[1], color='g', linestyle='--', label='95% CI Upper Bound')
plt.legend()
plt.show()

# Print the results
print("Bootstrap 95% Confidence Interval for R-squared:", confidence_interval)
```
