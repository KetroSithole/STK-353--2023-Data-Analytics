### Binomial distribution

```python
# Consider the experiment of a coin toss

# Parameters
num_flips = 100  # Number of coin flips
probability_heads = 0.5  # Probability of getting heads (0.5 for a fair coin)

# Simulate coin flips
coin_flips = np.random.binomial(1, probability_heads, num_flips)

# Count the number of heads and tails
num_heads = np.sum(coin_flips)
num_tails = num_flips - num_heads

print(f"Number of heads: {num_heads}")
print(f"Number of tails: {num_tails}")
```

```python
# Let's have a little fun and plot our experiment
import matplotlib.pyplot as plt

# Parameters
num_flips = 10
probability_heads = 0.5

# Simulate coin flips
coin_flips = np.random.binomial(1, probability_heads, num_flips)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(coin_flips, 'o-', label='Coin Flips')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Expected Value')
plt.xlabel('Flip Number')
plt.ylabel('Outcome (0 for Tails, 1 for Heads)')
plt.title('Simulated Coin Flips')
plt.legend()
plt.xticks(range(num_flips))
plt.ylim(-0.1, 1.1)
plt.grid()

plt.show()
```

## Gamma

```python
import scipy.special as sps

shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)
s = np.random.gamma(shape, scale, 1000)

count, bins, ignored = plt.hist(s, 50, density=True) 
y = bins**(shape-1)*(np.exp(-bins/scale) /  
                     (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')
plt.show()
```

### Weibull

```python
a = 5. # shape
s = np.random.weibull(a, 1000)

x = np.arange(1,100.)/50.
def weib(x,n,a):
    return (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)
count, bins, ignored = plt.hist(np.random.weibull(5.,1000))
x = np.arange(1,100.)/50.
scale = count.max()/weib(x, 1., 5.).max()
plt.plot(x, weib(x, 1., 5.)*scale)
plt.show()
```

### Pareto

```python
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 1000) + 1) * m

count, bins, _ = plt.hist(s, 100, density=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='r')
plt.show()
```

## Monte Carlo Integration

Monte Carlo Integration is a powerful numerical technique for approximating definite integrals using random sampling.

**Algorithm:**
1. Generate random points (x, y) within the interval [a, b] x [0, f_max], where f_max is an upper bound on f(x).

2. Count the number of points that lie under the curve.

3. Estimated integral = (Points under curve / Total points) * (b - a) * f_max


### Example 1
Imagine you want to estimate the value of the area of a circle using a technique that involves random sampling. You know that the area of a quarter circle with radius 1 is /4. Approximate the area of the circle using the Monte Carlo Integration method? 

Hint: 

_The area of a circle is given by $$A = \pi r^2,$$
 where $r$ is the radius. Thus the area of the circle is $\pi$._
 
 _Equation for unit circle is $$x^2 + y^2 = 1.$$_

```python
import numpy as np

def estimate_pi(num_samples):
    inside_circle = 0

    for _ in range(num_samples): # Step 1
        x = np.random.uniform(-1, 1) # Domain of X - unit circle
        y = np.random.uniform(-1, 1) # Domain of Y - unit circle
        
        if x**2 + y**2 <= 1: # Step 2
            inside_circle += 1  # Counts all points in circle
    
    pi_estimate = 4 * (inside_circle / num_samples)  # Step 3
    return pi_estimate

num_samples = 10000
estimated_pi = estimate_pi(num_samples)
print(f"Estimated value of : {estimated_pi}")
```

Visual representation of the points generated and the quarter circle. Blue points are inside the quarter circle, red points are outside. Our approximation of the integral considers a function of all the blue points vs the blue + red points.

```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(num_samples):
    inside_circle = 0

    # Lists to store points inside and outside the circle
    x_inside = []
    y_inside = []
    x_outside = []
    y_outside = []

    for _ in range(num_samples):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        
        if x**2 + y**2 <= 1:
            inside_circle += 1
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)
    
    pi_estimate = 4 * (inside_circle / num_samples)
    return pi_estimate, x_inside, y_inside, x_outside, y_outside

num_samples = 10000
estimated_pi, x_inside, y_inside, x_outside, y_outside = estimate_pi(num_samples)

# Visualization
plt.figure(figsize=(8, 8))
plt.scatter(x_inside, y_inside, color='blue', label='Inside Circle')
plt.scatter(x_outside, y_outside, color='red', label='Outside Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Monte Carlo Integration Visualization')
plt.legend()
plt.axis('equal')
plt.show()
```

## Probability Integral Transformation


Algorithm:
- Calculate the cumulative distribution function (CDF) $F(x)$ of the desired distribution.
- Generate a random sample U from a uniform distribution $U(0, 1)$.
- Apply the inverse of the CDF, $F^{(-1)}(u)$, to $U$ to obtain a sample from the desired distribution.


```python


# Generate random data from a non-uniform distribution (e.g., normal distribution)
np.random.seed(42)  # Setting seed for reproducibility
data = np.random.normal(loc=5, scale=2, size=1000)  # Normal distribution as an example

# Perform the Probability Integral Transformation
# Step 1: Sort the data
sorted_data = np.sort(data)

# Step 2: Calculate the cumulative distribution function (CDF)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Step 3: Apply the inverse of the CDF to transform the data into a uniform distribution
uniform_data = np.random.uniform(0, 1, len(sorted_data))  # Generate uniform random values
transformed_data = np.interp(uniform_data, cdf, sorted_data)  # Apply the inverse CDF

# Plot the original and transformed data
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, density=True, alpha=0.7, color='blue')
plt.title('Original Data')

plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=20, density=True, alpha=0.7, color='green')
plt.title('Transformed Data (Uniform Distribution)')

plt.tight_layout()
plt.show()
```

### Generating Exponential Samples
 Generate random samples from an exponential distribution using the Probability Integral Transformation. The exponential distribution is often used to model time between events.
Its probability density function (PDF) is given by $$f(x) = \lambda e^{(-\lambda x)}$$ for $x \ge 0$, where $\lambda$ is the rate parameter. The cumulative distribution function (CDF) is given as:
$$U = F(x) = 1 - e^{(-\lambda x)}$$
and thus, the inverse of the CDF is:
$$x = \frac{-ln(1-U)}{\lambda}$$
where $0 \le U < 1$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for exponential distribution
lambda_param = 0.6

# Generate uniform random samples
num_samples = 10000
uniform_samples = np.random.uniform(0, 1, num_samples)

# Apply probability integral transformation
exponential_samples = -np.log(1 - uniform_samples) / lambda_param

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(exponential_samples, bins=30, density=True, alpha=0.6, label='Generated Samples')
x = np.linspace(0, 10, 1000)
plt.plot(x, lambda_param * np.exp(-lambda_param * x), 'r', label='Exponential Distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Integral Transformation: Exponential Distribution')
plt.show()
```

#### Stock Return Modeling

Stock returns often follow non-normal distributions. Using the PIT we can generate random samples from the normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate uniform random samples
num_samples = 1000
uniform_samples = np.random.uniform(0, 1, num_samples)

# Apply probability integral transformation with symmetry around zero
## stats.norm.ppf - inverse of CDF function
desired_distribution_samples = stats.norm.ppf(2 * uniform_samples - 1)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(desired_distribution_samples, bins=30, density=True, alpha=0.6, label='Generated Samples')
x = np.linspace(-3, 3, 1000)
plt.plot(x, stats.norm.pdf(x), 'r', label='Standard Normal Distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Integral Transformation: Standard Normal Distribution')
plt.show()
```

## Bootstrap Method

Steps of the Bootstrap Algorithm
- Sample: Draw a sample of size n (with replacement) from the original dataset.
- Calculate Statistic: Calculate the desired statistic on the sampled data.
- Repeat: Repeat steps 1 and 2 a large number of times (e.g., B times) to create a distribution of the statistic.


#### Example: Estimating the Mean

Use the Bootstrap algorithm to estimate the mean of a dataset.

```python
import numpy as np

# Number of bootstrap iterations
def bootstrap(data, num_samples):
    sample_means = []
    
    for _ in range(num_samples):
        resampled_data = np.random.choice(data, size=len(data), replace=True) #Sample with replacement
        sample_mean = np.mean(resampled_data) #Compute mean of sample
        sample_means.append(sample_mean) #Add sample mean to vector of bootstrap means
    
    return sample_means

# Original dataset
data = np.array([3.2, 4.5, 5.1, 2.8, 6.2, 4.7, 3.9, 5.5, 4.1, 5.8])
# Number of bootstrap iterations
num_bootstrap_samples = 1000
# print(np.mean(data))
bootstrap_sample_means = bootstrap(data, num_bootstrap_samples)
# print(bootstrap_sample_means)
```

```python
import numpy as np

# Original dataset
data = np.array([12, 15, 18, 20, 24, 25, 28, 30, 32, 35])

# Number of bootstrap iterations
num_iterations = 10000

# Bootstrap resampling and mean estimation
bootstrap_means = [np.mean(np.random.choice(data, size=len(data), replace=True)) for _ in range(num_iterations)]

# Plotting
plt.hist(bootstrap_means, bins=30, density=True, alpha=0.6)
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Bootstrap Sampling Distribution of the Mean')
plt.show()
```

## Acceptance/Rejection Algorithm

 Steps of the Acceptance-Rejection Algorithm

- Choose a Proposal Distribution: Select a distribution that is easy to sample from and serves as a "proposal" for generating samples.
- Generate Samples: Draw random samples from the proposal distribution.
- Calculate Ratios: Calculate the ratio of the target distribution's probability to the proposal distribution's probability for each sample.
- Accept or Reject: Accept a sample with probability equal to the calculated ratio. Reject otherwise.

#### Example: Generating Exponential Samples
Use the Acceptance-Rejection Algorithm to generate random samples from an exponential distribution.
the target distribution is defined as $0.5 e^{-\frac{x}{2}}$, and the proposal distribution is the exponential distribution itself $e^{-x}$. The Acceptance-Rejection Algorithm aims to generate samples from the target distribution using the proposal distribution and the acceptance-rejection criteria.

```python
import numpy as np
import matplotlib.pyplot as plt

def target_distribution(x):
    return 0.5 * np.exp(-x / 2)

def proposal_distribution(x):
    return np.exp(-x)

def acceptance_rejection(num_samples):
    samples = []
    while len(samples) < num_samples:
        x = np.random.exponential(scale=1)
        u = np.random.uniform(0, 1)
        r = target_distribution(x) / (1.5 * proposal_distribution(x))
        if u <= r:
            samples.append(x)
    return samples

num_samples = 1000
samples = acceptance_rejection(num_samples)

# Plotting
plt.hist(samples, bins=30, density=True, alpha=0.6, label='Generated Samples')
x = np.linspace(0, 10, 1000)
plt.plot(x, target_distribution(x), 'r', label='Target Distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Acceptance-Rejection Algorithm: Exponential Distribution')
plt.show()
```
